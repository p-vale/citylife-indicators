```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
pop = pd.read_csv( '../data/pop.csv' )
```

```python
gender = pd.DataFrame(pop['gender'].value_counts( normalize = True ) * 100)
print(gender)

piechart = plt.pie( gender['gender'], 
                   startangle = 90,
                   labels = ['female', 'male'],
                   wedgeprops = {'edgecolor' : '#FFBD00',
                                 'facecolor': 'none'}
                )
piechart[0][1].set_hatch("///")
piechart[0][0].set_edgecolor('#00175B')

plt.show()
```
