```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fer = pd.read_csv( '../../data/fecondita.csv' )
```

```python
tasso = fer[ fer['Tipo dato'] == 'tasso di fecondità totale' ]
tasso = tasso.drop( columns = ['Tipo dato'])
tasso['Time'] = pd.to_datetime(tasso['Time'], format='%Y')
# print(type(tasso['Time'][0]))
```

```python
parents = fer[ fer['Tipo dato'] != 'tasso di fecondità totale' ]
parents.loc[ parents['Tipo dato'].str.contains('madri'), 'Tipo dato' ] = 'madri'
parents.loc[ parents['Tipo dato'].str.contains('padri'), 'Tipo dato' ] = 'padri'
```

```python
print(tasso.head())
print(parents.head())
```

```python
fig, ax = plt.subplots( figsize = (20, 10) )

plt.plot( 'Time', 
         'Value', 
         data = tasso, # up to here same as: ax.plot( tasso['Time'], tasso['Value'])
         color = 'skyblue', 
         marker = 'p', 
         markersize = 10
        )

ax.xaxis.grid( color = '#f0f0f0' )
ax.yaxis.grid( color = '#f0f0f0' )
# ax.set_facecolor('white')
ax.set_axisbelow(True)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(axis = u'both', 
               which = u'both',
               length = 0
              )

plt.savefig( 'fertility-tot.jpg', dpi = 150 )
```

```python
fig, ax = plt.subplots( figsize = (20, 10) )

plt.plot( 'Time', 
         'Value', 
         data = parents[parents['Tipo dato'].str.contains('madri')],
         color = '#0e77c7', 
         marker = 'h', 
         markersize = 10
        )

plt.plot( 'Time', 
         'Value', 
         data = parents[parents['Tipo dato'].str.contains('padri')],
         color = 'skyblue', 
         marker = 'p', 
         markersize = 10
        )

ax.xaxis.grid( color = '#f0f0f0' )
ax.yaxis.grid( color = '#f0f0f0' )
# ax.set_facecolor('white')
ax.set_axisbelow(True)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(axis = u'both', 
               which = u'both',
               length = 0
              )
plt.savefig( 'fertility-gen.jpg', dpi = 150 )
```
