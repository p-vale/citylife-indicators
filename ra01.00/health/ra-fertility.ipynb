```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
bg = '#11134A'
accent = '#202289'
txt = '#FAF9F1'
azzurro = '#6FCED9'
blu = '#7F8DDD'

mpl.rcParams['text.color'] = txt
mpl.rcParams['axes.labelcolor'] = txt
mpl.rcParams['xtick.color'] = txt
mpl.rcParams['ytick.color'] = txt
mpl.rcParams['axes.facecolor'] = bg
mpl.rcParams['axes.edgecolor'] = txt
```

```python
fer = pd.read_csv( '../../data/fecondita.csv' )
```

```python
tasso = fer[ fer['Tipo dato'] == 'tasso di fecondità totale' ]
tasso = tasso.drop( columns = ['Tipo dato'])
tasso['Time'] = pd.to_datetime(tasso['Time'], format='%Y')
# print(type(tasso['Time'][0]))
```

```python
parents = fer[ fer['Tipo dato'] != 'tasso di fecondità totale' ]
parents.loc[ parents['Tipo dato'].str.contains('madri'), 'Tipo dato' ] = 'madri'
parents.loc[ parents['Tipo dato'].str.contains('padri'), 'Tipo dato' ] = 'padri'
```

```python
print(tasso.head())
print(parents.head())
```

```python
fig, ax = plt.subplots( figsize = (20, 10) )

plt.plot( 'Time', 
         'Value', 
         data = tasso, # up to here same as: ax.plot( tasso['Time'], tasso['Value'])
         color = azzurro, 
         marker = 'p', 
         markersize = 10
        )

ax.xaxis.grid( color = '#f0f0f0' )
ax.yaxis.grid( color = '#f0f0f0' )
ax.set_axisbelow(True)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(axis = u'both', 
               which = u'both',
               length = 0
              )

fig.patch.set_facecolor(bg)
```

```python
fig, ax = plt.subplots( figsize = (20, 10) )

plt.plot( 'Time', 
         'Value', 
         data = parents[parents['Tipo dato'].str.contains('madri')],
         color = blu, 
         marker = 'h', 
         markersize = 10
        )

plt.plot( 'Time', 
         'Value', 
         data = parents[parents['Tipo dato'].str.contains('padri')],
         color = azzurro, 
         marker = 'p', 
         markersize = 10
        )

ax.xaxis.grid( color = '#f0f0f0' )
ax.yaxis.grid( color = '#f0f0f0' )
ax.set_axisbelow(True)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(axis = u'both', 
               which = u'both',
               length = 0
              )
fig.patch.set_facecolor(bg)
```
